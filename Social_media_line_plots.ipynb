```python
out['month_year'] = pd.to_datetime(out['month_year'])
out['Mean'] = out.groupby(['subcategory', out['month_year'].dt.year])['Count'].transform('mean')
out['Year'] = out['month_year'].dt.year
out['Kategorie'] = out['subcategory']

out

```

```python
# to do - calculate count_sum 


def plot_multiple_lineplot_charts(df, title):
    
    count_sum = df['Count'].sum()

    
    fig = px.line(df, x='Year', y='Mean', color='Kategorie',
                  labels={'Year': 'Jahr', 'Mean': 'Anzahl'},
                  title=f'Anzahl der Social - Media - Posts: {title}',
                  line_shape='linear',
                  color_discrete_sequence=PERMEA_COLOR_WHEEL_5_STATISTICS,
                  )

    fig.update_layout(
        template='plotly_white',
        yaxis_range=[0, 1.1 * df['Mean'].max()],
        xaxis=dict(
            tickmode='linear',
            dtick=1,
            automargin=True
        ),
        yaxis={'automargin': True},
        margin=dict(l=130, r=150, b=10, pad=30),
        autosize=False,
        width=20 * 37.8,  # Convert 20 cm to inches (1 cm = 37.8 inches)
        height=10 * 37.8,  # Convert 9 cm to inches (1 cm = 37.8 inches)
        legend=dict(font=dict(size=16))
    )


    fig.add_annotation(dict(font=dict(size=9),
                    x=-0.24,
                    y=1.15,
                    showarrow=False,
                    text=f"Basierend auf {count_sum} Social-Media-Konversationen aus öffentlich zugänglichen Foren <br>Deutschland, Januar 2013 – April 2023.",
                    textangle=0,
                    xanchor='left',
                    xref="paper",
                    yref="paper",
                    align="left"
))

    # fig.write_image(save_name+".png")
    # show plot
    fig.show()



plot_multiple_lineplot_charts(market_area_df, "Marktgebiet")
plot_multiple_lineplot_charts(buying_channel_df, "Einkaufskanal")
plot_multiple_lineplot_charts(main_brands_df, "Hauptmarken")
plot_multiple_lineplot_charts(website_df, "Website")
plot_multiple_lineplot_charts(rx_df, "RX vs. OTC")
plot_multiple_lineplot_charts(mode_of_administration, "Darreichungsform")
```

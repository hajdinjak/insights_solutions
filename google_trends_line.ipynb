```python
# remove less than 1 
def _sanitize(val):
    if val == "<1":
         val = "0"
    return val
```

```python
def shape(df):
    df.columns = df.columns.str.replace(':.*\)', '', regex=True)
    df.rename(columns={"Month": "Date"}, inplace=True)

    # Remove dates before x and after y
    df["Date"] = pd.to_datetime(df["Date"])
    df = df[df["Date"].dt.year >= 2013]
    df = df[df["Date"].dt.year <= 2023]
    #df = df[df["Date"].dt.month == 1]

    #Santize for bad values
    df = df.applymap(_sanitize)

    df = df.melt('Date', var_name='search_term',  value_name='interest')

    # Extract the year from the 'Month' column
    df['Year'] = df['Date'].dt.year

    # Calculate the mean yearly average for each category in 'search_term'
    mean_yearly_avg = df.groupby(['search_term', 'Year'])['interest'].mean().reset_index()

    # Merge the mean_yearly_avg with the original gtrends_df
    df = df.merge(mean_yearly_avg, on=['search_term', 'Year'], how='left', suffixes=('', '_year'))

    # Assign the mean yearly average to the 'interest_year' column
    df['interest_year'] = df['interest_year'].fillna(0)

    # Find the highest value when search_term is "Frauengesundheit"
    highest_value = df.loc[df["search_term"] == "Frauengesundheit", "interest_year"].max()

    # Calculate the scaling factor
    scaling_factor = 100 / highest_value

    # Multiply the values in interest_year column by the scaling factor
    df["interest_year"] *= scaling_factor


    return df




shaped_gtrends_df = shape(gtrends_df)
shaped_product_df = shape(product_df)
shaped_areas_df = shape(areas_df)



```
